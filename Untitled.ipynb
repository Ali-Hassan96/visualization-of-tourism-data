```python
labels = 'Air', 'Land', 'Sea'
sizes = [66.5, 33.2, 0.3]
explode = (0.1, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots(figsize=(10, 5))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("The precentge of type of travel", size=25)
plt.savefig("piechart.png", bbox_inches = 'tight')
plt.show()
```

```python
labels = 'Asia', 'Middle East', 'Other Countries', 'GCC', 'Europe', 'America', 'Africa'
sizes = [29, 17.9, 0.5, 37.2, 7.6, 1.5, 6.3]
explode = (0.1, 0, 0, 0, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

mpl.rcParams['font.size'] = 30
fig1, ax1 = plt.subplots(figsize=(15, 9))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=180, textprops={'fontsize': 20})
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('The precentge of trips by region', loc='left')
plt.show()
```
